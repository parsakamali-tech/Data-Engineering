```python
data = pd.read_csv('..\\data\\Sample_Superstore.csv', encoding='ISO-8859-1')
```

```python
data.columns.to_list
```

```python
#changing the columns name according to my db
data.columns=['row_id','order_id','order_date','ship_date','ship_mode','customer_id','customer_name','segment','country','city','state',
                    'postal_code','region','product_id','category','sub_category','product_name','sales','quantity','discount','profit']
```

```python
data
```

```python
#making customers table data
columns1 = ['customer_id','customer_name','segment','city','state','country']
df1 = data[columns1].drop_duplicates(subset='customer_id')#dropping duplicated data according to customer_id
df1.to_csv('..\\data\\customers.csv',index=False)
```

```python
#making products table data
columns2 = ['product_id','product_name','category','sub_category']
df2 = data[columns2].drop_duplicates(subset='product_id')#dropping duplicated data according to product_id
#df2.columns = columns2
df2.to_csv('..\\data\\products.csv',index=False)
```

```python
#making orders table data
columns3 = ['order_id','order_date','ship_date','customer_id','product_id','quantity','sales','profit']
df3 = data[columns3].drop_duplicates(subset='order_id')
df3.to_csv('..\\data\\orders.csv',index=False)
```
